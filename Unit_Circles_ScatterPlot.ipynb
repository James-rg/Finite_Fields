# Finite Fields

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = 19


x = [i for i in range(f)]
y = [i for i in range(f)]
xy = [[x1,y1] for x1 in x for y1 in y]
xy = np.array(xy)

plt.scatter(xy[:,0],xy[:,1],marker = "o", facecolors='none', edgecolors='b')
plt.axis('off')
plt.show()
```

```python

(xy[0,0]**2+xy[0,1]**2)%f==1
```

```python
circle = []

for i in xy:
  if((i[0]**2+i[1]**2)%f==1):
    circle.append(i.tolist())

```

```python
circle
```

```python

circle = np.array(circle)

plt.scatter(xy[:,0],xy[:,1],marker = "o", facecolors='none', edgecolors='b')
plt.scatter(circle[:,0],circle[:,1],marker = "o",color = 'b')
plt.axis('off')
plt.rcParams["figure.figsize"] = (200,200)
#plt.figure(figsize=(500, 500), dpi=800)
plt.show()
```
